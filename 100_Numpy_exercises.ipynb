#### 35. How to compute ((A+B)*(-A/2)) in place (without copy)? (★★☆)

````python
A=np.array(2.0)
B=np.array(3.0)
np.add(A, B, out=A)        # A = A + B
np.negative(A, out=A)     # A = -(A + B)
np.multiply(A, B, out=A)  # A = (A + B) * (-B)
np.divide(A, 2, out=A)    # A = ((A + B) * (-B)) / 2
````

#### 50. How to find the closest value (to a given scalar) in a vector? (★★☆)

````python
Z = np.array([1, 3, 7, 10])
x = 6

closest = Z[np.abs(Z - x).argmin()]
print(closest)
# Z - x → distance from x

# np.abs(...) → absolute distance

# .argmin() → index of smallest distance

# Z[...] → fetch the actual value
````
